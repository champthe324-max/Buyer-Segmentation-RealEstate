```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
buyer_df = pd.read_csv("../data/processed/final_buyer_dataset.csv")
```

```python
buyer_df.columns
```

```python
cluster_counts = buyer_df["Buyer_Persona"].value_counts()

plt.figure(figsize=(8,5))

cluster_counts.plot(kind="bar")

plt.title("Number of Buyers in Each Segment")
plt.xlabel("Buyer Segment")
plt.ylabel("Number of Buyers")
plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(8,5))

buyer_df.groupby("Buyer_Persona")["total_investment"].mean().sort_values().plot(
    kind="bar"
)

plt.title("Average Total Investment by Buyer Segment")
plt.xlabel("Buyer Segment")
plt.ylabel("Average Investment ($)")
plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(8,5))

buyer_df.groupby("Buyer_Persona")["avg_property_price"].mean().sort_values().plot(
    kind="bar"
)

plt.title("Average Property Price by Buyer Segment")
plt.xlabel("Buyer Segment")
plt.ylabel("Average Property Price ($)")
plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(8,5))

buyer_df.groupby("Buyer_Persona")["avg_property_area"].mean().sort_values().plot(
    kind="bar"
)

plt.title("Average Property Area by Buyer Segment")
plt.xlabel("Buyer Segment")
plt.ylabel("Area (sqft)")
plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(8,5))

buyer_df.groupby("Buyer_Persona")["satisfaction_score"].mean().sort_values().plot(
    kind="bar"
)

plt.title("Average Satisfaction Score")
plt.xlabel("Buyer Segment")
plt.ylabel("Satisfaction")
plt.xticks(rotation=20)

plt.show()
```

```python
loan = pd.crosstab(
    buyer_df["Buyer_Persona"],
    buyer_df["loan_applied"]
)

loan.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Loan Applied by Buyer Segment")
plt.xticks(rotation=20)

plt.show()
```

```python
purpose = pd.crosstab(
    buyer_df["Buyer_Persona"],
    buyer_df["acquisition_purpose"]
)

purpose.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Acquisition Purpose by Segment")
plt.xticks(rotation=20)

plt.show()
```

```python
ref = pd.crosstab(
    buyer_df["Buyer_Persona"],
    buyer_df["referral_channel"]
)

ref.plot(
    kind="bar",
    figsize=(9,5)
)

plt.title("Referral Channel by Buyer Segment")
plt.xticks(rotation=20)

plt.show()
```
