```python
# ============================================
# Import Required Libraries
# ============================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

pd.set_option("display.max_columns", None)

print("Libraries imported successfully.")
```

```python
# ============================================
# Load Datasets
# ============================================

clients_df = pd.read_csv("../data/raw/clients.csv")
properties_df = pd.read_csv("../data/raw/properties.csv")

print("Datasets loaded successfully.")
```

# 📝 Data Cleaning Strategy

Before performing any cleaning operations, we define a clear preprocessing strategy.

## Clients Dataset

- Check data types
- Convert `date_of_birth` into datetime
- Create Age feature
- Validate categorical values

## Properties Dataset

- Investigate missing values in `client_ref`
- Convert `transaction_date` into datetime
- Convert `sale_price` into numeric
- Validate property-related fields

## Final Goal

Prepare clean datasets for feature engineering and machine learning.

```python
# ============================================
# Check Data Types
# ============================================

print("Clients Dataset\n")
display(clients_df.dtypes)

print("\nProperties Dataset\n")
display(properties_df.dtypes)
```

```python
# ============================================
# Inspect Date Columns
# ============================================

print("Date of Birth Samples:")
display(clients_df["date_of_birth"].head(10))

print("\nTransaction Date Samples:")
display(properties_df["transaction_date"].head(10))
```

```python
# ============================================
# Inspect Sale Price Values
# ============================================

properties_df["sale_price"].head(10)
```

```python
# ============================================
# Convert Transaction Date to Datetime
# ============================================

properties_df["transaction_date"] = pd.to_datetime(
    properties_df["transaction_date"],
    format="%d-%m-%Y"
)

print("Transaction Date converted successfully.")
```

```python
properties_df["transaction_date"].head()
```

```python
properties_df["transaction_date"].head()
properties_df["transaction_date"].dtype
```

```python
# ============================================
# Check Sale Price Format
# ============================================

print("Unique Sale Price Samples:\n")

display(properties_df["sale_price"].sample(10, random_state=42))
```

```python
# ============================================
# Clean and Convert Sale Price
# ============================================

properties_df["sale_price"] = (
    properties_df["sale_price"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

print("Sale Price converted successfully.")
```

```python
properties_df["sale_price"].head()
properties_df["sale_price"].dtype
```

```python
# ============================================
# Convert Date of Birth to Datetime
# ============================================

clients_df["date_of_birth"] = pd.to_datetime(
    clients_df["date_of_birth"],
    format="mixed",
    dayfirst=False
)

print("Date of Birth converted successfully.")
```

```python
clients_df["date_of_birth"].dtype
clients_df["date_of_birth"].head()
```
